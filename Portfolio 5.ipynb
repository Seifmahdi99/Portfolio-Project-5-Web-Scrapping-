```python
# Import the AutoScraper Library
from autoscraper import AutoScraper
```

```python
# Get the Desired URL, and the info you want to scrape

FlipKart_URL = 'https://www.flipkart.com/search?q=Iphone'

Wanted_Info = ['₹77,900','APPLE iPhone 12 (Blue, 64 GB)',' 6,964 Ratings & 556 Reviews']

```

```python
# Doing the Automated Scraping

scraper = AutoScraper()

result = scraper.build(FlipKart_URL,Wanted_Info)

print(result)
```

```python
# Grouping of the Data

scraper.get_result_similar(FlipKart_URL,grouped=True)
```

```python
# Save Rules for other Pages

scraper.set_rule_aliases({'rule_mgj3': 'Price','rule_6j8r': 'Title'})

scraper.keep_rules(['rule_mgj3','rule_6j8r'])

scraper.save('FlipKart_Search')
```

```python
# Scrape a new Page using the rules

results = scraper.get_result_similar('https://www.flipkart.com/search?q=mi+mobiles',group_by_alias=True)

results
```
